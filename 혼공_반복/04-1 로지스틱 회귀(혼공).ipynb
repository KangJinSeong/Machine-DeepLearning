```python
'''
Date: 2022.08.30
Title: 
By: Kang Jin Seong
'''
''' ## 럭키백의 확률'''

'''### 데이터 준비 '''

import pandas as pd

fish = pd.read_csv('https://bit.ly/fish_csv_data')
fish.head() #pandas 데이터 프레임
```

```python
print(pd.unique(fish['Species']))
```

```python
fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy()
print(fish_input[:5])
```

```python
fish_target = fish['Species'].to_numpy()
print(fish_target[:5])
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(fish_input, fish_target, random_state = 42)
print(train_target.shape)

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier(n_neighbors= 3)
kn.fit(train_scaled, train_target)
print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))
```

```python
# 다중 분류에서 타켓값을 문자열 그대로 사용한다.
# 이때 문자열은 알파벳 순서로 저장된다.
print(kn.classes_)

# pd.unique() 고유값 추출 결과 값 과 다른 배열
print(pd.unique(fish['Species']))
```

```python
print(kn.predict(test_scaled[:5]))
```

```python
import numpy as np

proba = kn.predict_proba(test_scaled[:5])
print(np.round(proba, decimals = 4))
```

```python
# 로지스틱 회귀로 이진 분류 수행하기
# 불리언 인덱싱 사용하기
print(train_target.shape)
bream_smelt_indexes = (train_target == 'Bream') | (train_target == 'Smelt')
# bream_smelt_indexes.reshape(-1,1)
print(bream_smelt_indexes.shape)
print(train_scaled.shape)
train_bream_smelt = train_scaled[bream_smelt_indexes]
target_bream_smelt = train_target[bream_smelt_indexes]
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_bream_smelt, target_bream_smelt)

print(lr.predict(train_bream_smelt[:5]))
```

```python
print(lr.predict_proba(train_bream_smelt[:5]))
'''첫번째 열이 음성클래스(0), 두번쨰 열이 양성클래스(1) '''
```

```python
# 어떤게 음성 클래스 이고 양성클래스인지는 classes_속성에 저장
print(lr.classes_)
```

```python
# 로지스틱 회귀의 계수 확인
print(lr.coef_, lr.intercept_)
```

```python
# 선형회귀의 방정식의 Z 값을 계산
# 계산된 Z 값을 시그모이드 함수에 통과시켜 확률로 얻을 수 있다.
decisions = lr.decision_function(train_bream_smelt[:5])
print(decisions)
```

```python
# 파이썬의 시그모이드 함수 expit()
from scipy.special import expit
print(expit(decisions))
```

```python
'''
Date: 2022.09.02
Title: 
By: Kang Jin Seong
'''
# 로지스틱 회귀로 다중분류 수행하기

lr = LogisticRegression(C =20, max_iter = 1000)
lr.fit(train_scaled, train_target)
print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

```python
print(lr.predict(test_scaled[:5]))
```

```python
print(lr.classes_)
proba = lr.predict_proba(test_scaled[:5])
print(np.round(proba, decimals= 3))
```

```python
 print(lr.coef_.shape, lr.intercept_.shape)
```

```python
# 특성 5개(행), 타겟 7개(열)의 Z 값 계산
decision = lr.decision_function(test_scaled[:5])
print(np.round(decision, decimals= 2))
```

```python
from scipy.special import softmax

proba = softmax(decision, axis = 1)
print(np.round(proba, decimals=3))
```
